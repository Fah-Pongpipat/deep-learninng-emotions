```python
import os
from PIL import Image

# ฟังก์ชันในการดึงข้อมูลจากรูปภาพ
def get_image_info(image_path):
    try:
        with Image.open(image_path) as img:
            width, height = img.size
            mode = img.mode
            channels = 1 if mode == 'L' else 3 if mode in ['RGB', 'RGBA'] else 0
            print(f"Image: {os.path.basename(image_path)}, Width: {width}, Height: {height}, Mode: {mode}, Channels: {channels}")
            return width, height, mode, channels
    except Exception as e:
        print(f"Error loading image {image_path}: {e}")
        return None, None, None, None

# ฟังก์ชันหลักเพื่อทดสอบรูปภาพในโฟลเดอร์
def analyze_images_in_folder(folder_path):
    for filename in os.listdir(folder_path):
        if filename.endswith(('.png', '.jpg', '.jpeg')):  # ประเภทไฟล์ที่ต้องการ
            image_path = os.path.join(folder_path, filename)
            get_image_info(image_path)

# ป้อน path ของโฟลเดอร์ที่มีรูปภาพที่ต้องการตรวจสอบ
folder_path = 'datasets/train/angry'  # เปลี่ยนเป็น path ที่ถูกต้อง
analyze_images_in_folder(folder_path)
```
